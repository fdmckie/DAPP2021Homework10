#### 9. Create a crosstab (cross-tabulation table) to answer the following questions:

#### 11. Use the groupby function to determine the average age of passengers by passenger class.

#### 12. Locate all the passengers who embarked at "Queenstown".

#### 13. Locate the teenagers who were onboard the Titanic - passengers ages 13 to 19. 

```python
# using the "and" logical operator & the "loc" method, look at the "age" column to find the rows with teenagers 
# (at least age 13 & less than age 20)...

titanic.loc[(titanic["age"] >= 13.0) & (titanic["age"] < 20.0)]
```

```python
# 131 passengers were teenagers
```

#### 14. Locate the passengers who were <i>either</i> in second or third class.

```python
# using the "or" logical operator & the "loc" method, look at the "pclass" column to find all of the 
# 2nd & 3rd class passengers...


titanic.loc[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)]
```

```python
# 986 passengers were in 2nd class or 3rd class...
```

```python
## Question 1: how many people survived? how many died?

# use the "value_counts" method to discover how many people survived & died (look at "survived" method)...

titanic["survived"].value_counts()
```

```python
# with the value of "0" (did not survive): 809 people...
# with the value of "1" (survived): 500 people...
```

```python
## Question 2: from the people that survived & died, from which passenger class?

# use the "crosstab" method to discover the number of passengers that survived or died (look at "survived" column) 
# by passenger class (look at "pclass" column)...

pd.crosstab(titanic["survived"], titanic["pclass"], margins=True)
```

```python
# from passenger class 1: 
# "survived" value of "0" (did not survive): 123
# "survived" value of "1" (survived): 200

# from passenger class 2: 
# "survived" value of "0" (did not survive): 158
# "survived" value of "1" (survived): 119

# from passenger class 3: 
# "survived" value of "0" (did not survive): 528
# "survived" value of "1" (survived): 181
```

```python
## Question 3: from the people that survived & died, from where they embarked?

# use the "crosstab" method to discover the number of passengers that survived or died (look at "survived" column) 
# by boarding port (look at "embarked" column)...
#C = Cherbourg, Q = Queenstown, S = Southampton

pd.crosstab(titanic["survived"], titanic["embarked"], margins=True)
```

```python
# from boarding in Cherbourg ("C"): 
# "survived" value of "0" (did not survive): 120
# "survived" value of "1" (survived): 150

# from boarding in Queenstown ("Q"): 
# "survived" value of "0" (did not survive): 79
# "survived" value of "1" (survived): 44

# from boarding in Southhampton ("S"):
# "survived" value of "0" (did not survive): 610
# "survived" value of "1" (survived): 304
```

```python
## Question 4: from the people that survived & died, the breakdown by sex?

# use the "crosstab" method to discover the number of passengers that survived or died (look at "survived" column) 
# by sex (look at "sex" column)...

pd.crosstab(titanic["survived"], titanic["sex"], margins=True)
```

```python
# by male: 
# "survived" value of "0" (did not survive): 682
# "survived" value of "1" (survived): 161

# by female: 
# "survived" value of "0" (did not survive): 127
# "survived" value of "1" (survived): 339
```
